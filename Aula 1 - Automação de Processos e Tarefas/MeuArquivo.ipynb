# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

Comandos pyautogui: https://pyautogui.readthedocs.io/en/latest/quickstart.html

```python
import pyautogui as pag
import pyperclip as ppc
import time

pag.PAUSE = 1 # Pausa que vai ser dada entre cada instrução

# Passo 1: Entrar no sistema (no nosso caso, entrar no link do drive)
pag.press("win")
pag.write("chrome")
pag.press("enter")
ppc.copy("https://drive.google.com/drive/folders/14oLE59U1RqyRqlBbKpsyymW-mitvbtoh")
pag.hotkey("ctrl", "v")
pag.press("enter")
time.sleep(4) # tempo para abrir a página

# Passo 2: Baixar o arquivo de dados
pag.click(x=465, y=450)
pag.click(x=1657, y=237)
time.sleep(1) 
pag.click(x=1474, y=741)

time.sleep(3) # tempo para carregar o arquivo
pag.press("enter")
time.sleep(3) # tempo fazendo o download

# Passo 3: Ler o arquivo para pegar o indicadores (Faturamento e Quantidade de Produtos)
import pandas as pd
from IPython.display import display

tabela = pd.read_excel(r"C:\Users\luiza\Desktop\Vendas - Dez.xlsx") # o "r" significa "row string" e faz com que o programa desconsidere caracteres especiais 
display(tabela)
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()

# Passo 4: Entrar no Gmail
pag.hotkey("ctrl", "t")
ppc.copy("https://mail.google.com/mail/u/0/#inbox")
pag.hotkey("ctrl", "v")
pag.press("enter")
time.sleep(3) # tempo para abrir a página do gmail

# Passo 5 Enviar um email para a diretoria (no caso, eu mesma)
pag.click(x=80, y=246)
ppc.copy("luizablaquini@gmail.com")
pag.hotkey("ctrl", "v")
pag.press("tab") # pula para o campo de assunto
ppc.copy("Relatório mensal de Vendas [DEZEMBRO]")
pag.hotkey("ctrl", "v")
pag.press("tab") # pula para o campo da mensagem

texto = f"""Bom dia

Relativo ao mês de DEZEMBRO:

O faturamento do mês foi de: R${faturamento:,.2f}
A quantidade de produtos vendida foi de: {quantidade:,}

Abs
Att: Luiza B Laquini
"""

ppc.copy(texto)
pag.hotkey("ctrl", "v")

pag.press("tab") # pula para o botão de enviar
pag.press("enter")
```
